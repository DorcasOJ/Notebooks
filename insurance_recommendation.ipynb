```python
def melt_and_clean(df):
    #fill values
    df = df.fillna(df.join_date.mode().iloc[0])
    #time stramp
    df.join_date = df.join_date.astype('str')
    df.join_date = [time.mktime(datetime.datetime.strptime(i, '%d/%m/%Y').timetuple()) for i in df.join_date]
    
    #melt
    pcode = list(df.columns[-21:])
    idt = list(df.columns[:-21])
    df = pd.melt(df, id_vars = idt, value_vars = pcode, var_name= 'ID X PCODE', value_name='Label')
    pcode_join = list(df.ID + ' X '+ df['ID X PCODE'])
    df['ID X PCODE'] = pcode_join
    
    return df
```

```python
train_new = melt_and_clean(train)
train_new.head()
```

```python
test_new = melt_and_clean(test)
test_new.head()
```

```python
def dummy(df):
    #dummies
    sex = pd.get_dummies(df.sex)
    marital = pd.get_dummies(df.marital_status)
    branch = pd.get_dummies(df.branch_code)
    occup_cat = pd.get_dummies(df.occupation_category_code)
    
    #df.birth_year = df.birth_year.astype('datetime64[ns]')
    #concat and drop
    col_code=[i for i in df.columns if i.endswith('_code')]
    dum=['sex','marital_status']
    cols = col_code + dum
    df = df.drop(cols, axis=1)
    df = pd.concat([df, sex, marital, branch, occup_cat], axis=1)
    
    return df
    
```

```python
train_update = dummy(train_new)
train_update.head()
```

```python
test_update = dummy(test_new)
test_update.head()
```
